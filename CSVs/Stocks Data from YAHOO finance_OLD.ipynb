```python
import requests
import pandas as pd
import json
```

## TESLA

```python
url = "https://apidojo-yahoo-finance-v1.p.rapidapi.com/stock/v3/get-historical-data"
```

```python
querystring = {"symbol":"TSLA","region":"US"}
```

```python
headers = {
    'x-rapidapi-key': "cf7c79b3ffmshb02951f77e9bd67p15b844jsna29f69c36227",
    'x-rapidapi-host': "apidojo-yahoo-finance-v1.p.rapidapi.com"
    }
```

```python
response = requests.request("GET", url, headers=headers, params=querystring)
```

```python
print(response.text)
```

```python
print(response.status_code)
```

```python
# Create a function that prints out the JSON formatted string

def print_json(response):
    word = json.dumps(response, sort_keys=True, indent=4)
    
    print(word)

    
print_json(response.json())  
```

```python
data=response.json()
```

```python
columns=[x for x in data["prices"]]
columns
```

```python
df=pd.DataFrame(columns)
df
```

```python
#save the list as a csv
df.to_csv("TESLA STOCK DATA_18_01_21.csv")
```
